# 02 Exploratory Analysis

This notebook explores the AWGC phase-evolution datasets using basic visualization.

The goal is to understand how sintering temperature and SBF soaking time influence phase composition and bioactivity-related trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

ROOT_DIR = Path('..')
RAW_DATA_DIR = ROOT_DIR / 'data' / 'raw'
FIGURES_DIR = ROOT_DIR / 'figures'
FIGURES_DIR.mkdir(parents=True, exist_ok=True)

initial_df = pd.read_csv(RAW_DATA_DIR / 'initial_phase_composition.csv')
sbf_df = pd.read_csv(RAW_DATA_DIR / 'sbf_phase_evolution.csv')

initial_df.head(), sbf_df.head()
```

## Initial Phase Composition

This plot shows how the initial phase composition changes with sintering temperature.

```python
plt.figure(figsize=(8, 5))

plt.plot(initial_df['temperature_C'], initial_df['amorphous_pct'], marker='o', label='Amorphous')
plt.plot(initial_df['temperature_C'], initial_df['wollastonite_pct'], marker='o', label='Wollastonite')
plt.plot(initial_df['temperature_C'], initial_df['hydroxyapatite_pct'], marker='o', label='Hydroxyapatite')
plt.plot(initial_df['temperature_C'], initial_df['whitlockite_pct'], marker='o', label='Whitlockite')

plt.xlabel('Sintering temperature (°C)')
plt.ylabel('Phase fraction (%)')
plt.title('Initial Phase Composition of AWGC')
plt.legend()
plt.tight_layout()

plt.savefig(FIGURES_DIR / 'initial_phase_composition.png', dpi=300)
plt.show()
```

## Hydroxyapatite Evolution During SBF Immersion

Hydroxyapatite formation is used as a key indicator of bioactivity.

```python
plt.figure(figsize=(8, 5))

for temperature in sorted(sbf_df['temperature_C'].unique()):
    subset = sbf_df[sbf_df['temperature_C'] == temperature]
    plt.plot(
        subset['soaking_day'],
        subset['hydroxyapatite_pct'],
        marker='o',
        label=f'{temperature} °C'
    )

plt.xlabel('SBF soaking time (days)')
plt.ylabel('Hydroxyapatite fraction (%)')
plt.title('Hydroxyapatite Evolution During SBF Immersion')
plt.legend(title='Sintering temperature')
plt.tight_layout()

plt.savefig(FIGURES_DIR / 'sbf_hydroxyapatite_evolution.png', dpi=300)
plt.show()
```

## Wollastonite Evolution During SBF Immersion

Wollastonite is important because it contributes to the bioactive and mechanically reinforcing behavior of AWGC systems.

```python
plt.figure(figsize=(8, 5))

for temperature in sorted(sbf_df['temperature_C'].unique()):
    subset = sbf_df[sbf_df['temperature_C'] == temperature]
    plt.plot(
        subset['soaking_day'],
        subset['wollastonite_pct'],
        marker='o',
        label=f'{temperature} °C'
    )

plt.xlabel('SBF soaking time (days)')
plt.ylabel('Wollastonite fraction (%)')
plt.title('Wollastonite Evolution During SBF Immersion')
plt.legend(title='Sintering temperature')
plt.tight_layout()

plt.savefig(FIGURES_DIR / 'sbf_wollastonite_evolution.png', dpi=300)
plt.show()
```
